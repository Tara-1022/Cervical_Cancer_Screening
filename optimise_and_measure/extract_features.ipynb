```python
path = r"...\classification_model.pt"
```

```python
image = r"...\image.jpg"
```

```python
import cv2
import numpy as np
import os
import pandas as pd
```

```python
import torch
from ultralytics import YOLO

model = YOLO(path)

def get_feature_map(image_path):
    feature_maps = []
    def hook(module, input, output):
        feature_maps.append(output)
    layer = dict(model.named_modules())['model.model.9.drop']
    layer.register_forward_hook(hook)

    with torch.no_grad():
        outputs = model(cv2.imread(image_path))

    return feature_maps[0].numpy()
tensor = get_feature_map(image)
print(tensor)
print(tensor.shape)
```

```python
data_path = r"path_to_data"
VAL_PATHS = [data_path + "\\val\\" +  folder for folder in list(os.walk(data_path + "\\val"))[0][1]]
print(VAL_PATHS)
TRAIN_PATHS = [data_path + "\\train\\" +  folder for folder in list(os.walk(data_path + "\\train"))[0][1]]
print(TRAIN_PATHS)
```

```python
vals = list()
for p in VAL_PATHS:
    i = 0
    filenames = [i for i in os.listdir(p) if i.endswith(('.png', '.jpg', '.jpeg'))]
    for image in filenames:
        image_path = os.path.join(p, image)
        print(image_path)
        features = get_feature_map(image_path)
        features = np.append(features, ('CIN1-' if 'CIN1-' in p else 'CIN2+', image))
        print(features)
        vals.append(features)
        i += 1
        # if i == 2:
        #     break
    print(i, " done in ", p)
print(vals)
```

```python
validation = pd.DataFrame(vals)
validation
```

```python
validation.to_csv("Features_validation.csv")
```

```python
print(TRAIN_PATHS)
trains = list()
for p in TRAIN_PATHS:
    i = 0
    filenames = [i for i in os.listdir(p) if i.endswith(('.png', '.jpg', '.jpeg'))]
    for image in filenames:
        image_path = os.path.join(p, image)
        print(image_path)
        features = get_feature_map(image_path)
        features = np.append(features, ('CIN1-' if 'CIN1-' in p else 'CIN2+', image))
        print(features)
        trains.append(features)
        i += 1
        # if i == 2:
        #     break
    print(i, " done in ", p)
print(trains)
```

```python
training = pd.DataFrame(trains)
training
```
