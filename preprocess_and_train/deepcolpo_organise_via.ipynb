```python
import pandas as pd
import shutil
```

# Read 

```python
via_meta = pd.read_excel(r"path_to\VIA Cases Meta data.xlsx" )

via_meta = via_meta[["CaseNumber", "CaseID", "VIA"]]
via_meta.head(2)
```

```python
via_image_dat = pd.read_excel(r"path_to\Cases - Images.xlsx")
via_image_dat= via_image_dat[["CaseNumber", "File", "Type"]]
via_image_dat.head(2)
```

```python
def find_matches(row, df2):
    matches = df2[df2['File'].str.startswith(row['CaseID'])]
    return matches
```

```python
results = []

for _, row in via_meta.iterrows():
    matches = find_matches(row, via_image_dat)
    for _, match in matches.iterrows():
        print("FALSE:" + str(row['CaseNumber'] ) if row['CaseNumber'] != match['CaseNumber'] else "", end = "")
        results.append({
            'File': match['File'],
            'CaseID': row['CaseID'],
            'CaseNumber': row['CaseNumber'],
            'Type': match['Type'],
            'Class': row['VIA']
        })

final_via_set = pd.DataFrame(results)
```

# Move data

```python
def copy_images(from_to):
    c = 0
    for fr, to in from_to:
        try:
            shutil.copy(fr, to)
        except:
            print("ERROR:", fr, to)
        c +=1
        print(c, "done")
```

```python
BASE_PATH_FROM = r"path_to\VIA"
BASE_PATH_TO = r"path_to_folders"
```

## Manage special cases first

```python
ABNORMAL_CLASSES  = [i for i in via_meta["VIA"].unique() if any(w in i.lower() for w in ["positive", "cancer"])]
ABNORMAL_CLASSES
```

```python
# generate image list for Cancer folder
abnormal_path_to = BASE_PATH_TO + "\\Abnormal\\"
abnormal_from_to = []
for class_ in ABNORMAL_CLASSES:
    rows = final_via_set[final_via_set['Class'] == class_]
    paths = [
        ( BASE_PATH_FROM + "\\Case {:03}\\{}".format(
            row["CaseNumber"],
            row["File"]
            ),
          abnormal_path_to + row['File'])
            for _, row in rows.iterrows()]
    abnormal_from_to.extend(paths)
abnormal_from_to[:2]
```

```python
copy_images(abnormal_from_to)
```

## Move rest of images

```python
NORMAL_CLASSES = ["Negative"]
```

```python
all_from_to = list()
for class_ in NORMAL_CLASSES:
    rows = final_via_set[final_via_set['Class'] == class_]
    paths = [
        ( BASE_PATH_FROM + "\\Case {:03}\\{}".format(
            row["CaseNumber"],
            row["File"]
            ),
          BASE_PATH_TO + "\\Normal\\" + row['File'])
            for _, row in rows.iterrows()]
    all_from_to.extend(paths)
all_from_to[:2]
    
```

```python
copy_images(all_from_to)
```
